## ***emporiko***

```python
name = "emporiko"
```

```python
# Load the shapefile
df = pd.read_csv(f"/content/drive/My Drive/Diplomatiki/geocoded-initial/geocoded-" + name +".csv")
print(f"samples: ", len(df))
df.head()
```

```python
df_clean = df.drop_duplicates(subset=['full addre'], keep='first')
print(len(df_clean))
```

```python
df_clean[['latitude', 'longitude']] = df_clean['latlong'].str.split(',', expand=True)

# Convert to float
df_clean['latitude'] = df_clean['latitude'].astype(float)
df_clean['longitude'] = df_clean['longitude'].astype(float)
df_clean.head()
```

```python
df_clean = df_clean[['full addre', 'city', 'country', 'latitude', 'longitude', 'active yea', 'successful', 'predicted', 'pred_suc', 'osm_id', 'display_na', 'category', 'type']]
df_clean.rename(columns={'full addre': 'full adress'}, inplace=True)
df_clean.rename(columns={'active yea': 'active years'}, inplace=True)
df_clean.rename(columns={'successful': 'success'}, inplace=True)
df_clean.rename(columns={'predicted': 'predicted success prob'}, inplace=True)
df_clean.rename(columns={'pred_suc': 'predicted success'}, inplace=True)
df_clean.head()
```

```python
df_clean.to_csv(f'geocoded-' + name + '.csv', index=False)
```
