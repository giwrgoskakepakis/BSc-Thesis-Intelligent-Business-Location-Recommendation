# ***fine-tune attempt***

## Model Class

```python
import torch
from transformers import BartForSequenceClassification

class BartForProbabilityRegression(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.model = BartForSequenceClassification.from_pretrained("facebook/bart-large-mnli", num_labels=1, ignore_mismatched_sizes=True)
        self.model.classifier = torch.nn.Linear(self.model.config.hidden_size, 1)  # Single neuron for regression
        self.sigmoid = torch.nn.Sigmoid()  # Ensure output is in [0,1]
        self.loss_fn = torch.nn.MSELoss()

    def forward(self, input_ids, attention_mask, labels=None):
        outputs = self.model(input_ids, attention_mask=attention_mask)
        logits = self.model.classifier(outputs[0])  # Linear layer output
        probability = self.sigmoid(logits)  # Apply sigmoid to keep in [0,1]

        loss = None
        if labels is not None:
            loss = self.loss_fn(probability.squeeze(-1), labels.float())  # Ensure correct shape

        return {"loss": loss, "logits": probability}
```

## Load the data

```python
departments = ['ΥΠΗΡΕΣΙΩΝ', 'ΕΜΠΟΡΙΚΟ', 'ΜΕΤΑΠΟΙΗΤΙΚΟ', 'ΤΟΥΡΙΣΤΙΚΟ']
```

```python
fine_tuning_dfs = []

for department in departments:

    # load llm prompts
    llm_df = pd.read_csv(f"/content/drive/My Drive/Diplomatiki/LLM_prompt_data/data_LLM_" + department + ".csv")

    # load real data
    exel_df = pd.read_csv(f"/content/drive/My Drive/Diplomatiki/Success Probability - Real/data_" + department + ".csv")
    exel_df = exel_df[['active years scaled']]

    # make llm df
    fine_tuning_df = llm_df[['text_feature']].rename(columns={'text_feature': 'text'})
    fine_tuning_df['labels'] = exel_df['active years scaled']
    fine_tuning_dfs.append(fine_tuning_df)
    print(fine_tuning_df.head())
```

```python
from sklearn.model_selection import train_test_split

final_train_df = pd.DataFrame()
final_test_df = pd.DataFrame()

for df in fine_tuning_dfs:

    # slplit the df
    train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

    # concat the dfs
    final_train_df = pd.concat([final_train_df, train_df], ignore_index=True)
    final_test_df = pd.concat([final_test_df, test_df], ignore_index=True)

print("Final Training Data:")
print(final_train_df.head())
print("Final Test Data:")
print(final_test_df.head())
```

```python
from datasets import Dataset

dataset_train = Dataset.from_pandas(final_train_df)
dataset_test = Dataset.from_pandas(final_test_df)
```

## Load the model

```python
# Tokenize text columns
def tokenize_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True)
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("facebook/bart-large-mnli")
```

```python
tokenized_train = dataset_train.map(tokenize_function, batched=True)
tokenized_test = dataset_test.map(tokenize_function, batched=True)
```

## Train the model

```python
from transformers import Trainer, TrainingArguments

# Training arguments
training_args = TrainingArguments(
    output_dir="./results",
    num_train_epochs=3,
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    evaluation_strategy="epoch",
)
```

```python
trainer = Trainer(
    model=BartForProbabilityRegression(),
    args=training_args,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_test
)
```

```python
trainer.train()
```
