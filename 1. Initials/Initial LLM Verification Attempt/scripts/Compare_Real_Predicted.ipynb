# **initials**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.metrics import log_loss, confusion_matrix, roc_curve, roc_auc_score, classification_report, brier_score_loss
from sklearn.calibration import calibration_curve
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

# ***Department: ΥΠΗΡΕΣΙΩΝ***

```python
department="ΥΠΗΡΕΣΙΩΝ"
```

```python
df = pd.read_csv(f"/content/drive/My Drive/Diplomatiki/Success Probability - Merged/merged_data_" +department +".csv")
df['predicted_success'] = (df['predicted success probability'] >= 0.5).astype(int)
print(f"Total samples: ", len(df))
df.head()
```

```python
# make confusion matrix
cm = confusion_matrix(df['successful'], df['predicted_success'])

sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Predicted 0', 'Predicted 1'], yticklabels=['Actual 0', 'Actual 1'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Compute ROC curve
fpr, tpr, thresholds = roc_curve(df['successful'], df['predicted success probability'])

# Plot ROC curve
plt.figure(figsize=(6,6))
plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')  # Diagonal line (random classifier)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```

```python
# classification report
report = classification_report(df['successful'], df['predicted_success'])
print(report)
```

```python
# log loss
logloss = log_loss(df['successful'], df['predicted success probability'])
print(f'Log Loss: {logloss:.4f}')
```

```python
# brier score
brier = brier_score_loss(df['successful'], df['predicted success probability'])
print(f'Brier Score: {brier:.4f}')
```

```python
# Compute calibration curve
prob_true, prob_pred = calibration_curve(df['successful'], df['predicted success probability'], n_bins=10)

# Plot calibration curve
plt.figure(figsize=(6,6))
plt.plot(prob_pred, prob_true, marker='o', label='Model Calibration')
plt.plot([0, 1], [0, 1], linestyle='--', label='Perfect Calibration')
plt.xlabel('Predicted Probability')
plt.ylabel('Fraction of Positives')
plt.title('Calibration Curve')
plt.legend()
plt.show()
```
