```python
import pandas as pd
import folium
import geopandas as gpd
```

```python
# Check current CRS
grid = gpd.read_file("C:/Users/Giorgos/Desktop/HMMY/10ο Εξάμηνο/Διπλωματική/2. Datasets/ELSTAT/ΔΗΜΟΤΙΚΕΣ ΚΟΙΝΟΤΗΤΕΣ - shapefiles/grid/magnesia-grid.shp")
print("Original CRS:", grid.crs)
```

```python
# Reproject to WGS 84 (EPSG:4326)
grid = grid.to_crs(epsg=4326)
```

```python
# Compute centroids
grid['centroid_lon'] = grid.geometry.centroid.x
grid['centroid_lat'] = grid.geometry.centroid.y

# Save to CSV
grid.to_csv("grids2.csv", index=False)
```

```python
# Load the CSV
df = pd.read_csv("grids2.csv")
```

```python
df.head()
```

```python
# Create base map centered around Magnesia
m = folium.Map(location=[df['centroid_lat'].mean(), df['centroid_lon'].mean()], zoom_start=11)

# Add each grid centroid as a marker or circle
for _, row in df.iterrows():
    folium.CircleMarker(
        location=[row['centroid_lat'], row['centroid_lon']],
        radius=0.5,
        color='blue',
        fill=True,
        fill_opacity=0.6
    ).add_to(m)
```

```python
# Save and display
m.save("magnesia_grid_centroids.html")
```
