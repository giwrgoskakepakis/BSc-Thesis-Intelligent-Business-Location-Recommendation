## **Multiple API Calls - Actual API Calls**

Okay, now that everything works, i can actually perform the API Calls for the business entries. I am going to save ***each grid call in a different .json file***

```python
import requests
import json
import time

#API_KEY = "fsq3J66UHgBF5qzppjl1FrV78onTjKuFfLsfy"

# Grid config
center_lat = 39.36635948927823 
center_lng = 22.951358477687226
lat_step = 0.01
lng_step = 0.01

grid_size = 5         # Creates a 5x5 grid (25 requests total)
radius = 1000         # Search radius in meters
limit = 50            # Max allowed by Foursquare

file_counter = 1

for i in range(-grid_size // 2, grid_size // 2 + 1):
    for j in range(-grid_size // 2, grid_size // 2 + 1):

        # grid center
        lat = center_lat + (i * lat_step)
        lng = center_lng + (j * lng_step)

        # set request parameters
        url = "https://api.foursquare.com/v3/places/search"
        headers = {
            "Accept": "application/json",
            "Authorization": "fsq3J66UHgBF5qzppjl1FrV78onTjKuFfLsfy/5sSAz+lXQ="
        }
        params = {
            "ll": f"{lat},{lng}",
            "radius": radius,
            "limit": limit,
            "fields": "fsq_id,name,geocodes,location,categories,chains,related_places,timezone"
        }

        print(f"Requesting places at {lat:.5f}, {lng:.5f}...")

        try:

            # make the request
            response = requests.get(url, headers=headers, params=params)
            response.raise_for_status()
            data = response.json()

            places = data.get("results", [])

            # store results
            if places:
                filename = f"places_{file_counter}.json"
                with open(filename, "w", encoding="utf-8") as f:
                    json.dump(places, f, indent=2, ensure_ascii=False)
                print(f"Saved {len(places)} places to {filename}")
                file_counter += 1
            else:
                print("No places found here.")

        except requests.RequestException as e:
            print(f"Request failed: {e}")

        # Optional: Be kind to the API
        time.sleep(0.5)

print("Done collecting places.")
```
