# ***Forsquare - Xrysos Odigos - Greek Catalog***

## ***Initials***

```python
import pandas as pd
```

## ***Load the Data***

```python
df1 = pd.read_csv("API Calls/Fourthsquare/forthsquare_data.csv")
df2 = pd.read_csv("Web Scraping/Χρυσός Οδηγός/xrysos-odigos-volos-ratings.csv")
df3 = pd.read_csv("Web Scraping/Greek Catalog/greek-catalog-volos.csv")

df1['Source'] = 'foursquare'
df2['Source'] = 'xrysos odigos'
df3['Source'] = 'greek catalog'
```

```python
df1.drop(columns=['category_id'], inplace=True)
df1.head()
```

```python
df2.head()
```

```python
df3.head()
```

## ***Rename Columns***

We need to assign the same names to the corresponding colunms in order to merge the 3 dfs

```python
df1 = df1.rename(columns={
    'name': 'Name',
    'category_name': 'Category',
    'lat': 'Latitude',
    'long': 'Longitude',
    'postal_code': 'ZIP',
    'address': 'Address',
    'country': 'Country',
    'city': 'City',
    'rating' : 'Rating',
    'reviews' : 'Reviews',
    'ZIP' : 'Postal Code',
})

df1.head()
```

```python
df2 = df2.rename(columns={
    'businessName': 'Name',
    'category': 'Category',
    'number_of_ratings': 'Number_of_ratings',
    'rating': 'Rating',
    'Number_of_ratings': 'Reviews',
    'ZIP' : 'Postal Code',
    'description': 'Description',
})

df2.head()
```

```python
df3 = df3.rename(columns={
    'name': 'Name',
    'category': 'Category',
    'number_of_ratings': 'Reviews',
    'rating' : 'Rating',
    'ZIP' : 'Postal Code',
})

df3.head()
```

```python
# Ensure ZIP codes are string type to preserve leading zeros
for df in [df1, df2, df3]:
    df['Postal Code'] = df['Postal Code'].astype(str).str.strip()
```

## ***Merge the Data***

```python
combined_df = pd.concat([df1, df2, df3], ignore_index=True)
```

```python
combined_df.head()
```

## ***Save the Data***

```python
combined_df.to_csv("forsquare_xrysosOdigos_greekCatalog.csv", index=False)
```

```python
print(f"Lengths: {len(df1)} + {len(df2)} + {len(df3)} = {len(df1) + len(df2) + len(df3)}")
print(len(combined_df))
```
