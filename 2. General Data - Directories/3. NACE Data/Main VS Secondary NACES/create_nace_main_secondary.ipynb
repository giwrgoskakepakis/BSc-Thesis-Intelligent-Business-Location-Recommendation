### ***Percentage of total entries***

```python
import pandas as pd

def count_main_nace_businesses(business_df, main_nace_df):
    
    # Get the set of main NACE codes (as strings to avoid type mismatches)
    main_naces = set(main_nace_df['NACE'].astype(str))

    # Ensure business NACE codes are strings too
    business_df['NACE Code'] = business_df['NACE Code'].astype(str)

    # Count how many businesses belong to main NACEs
    matching = business_df[business_df['NACE Code'].isin(main_naces)]
    
    return len(matching), len(business_df), matching
```

```python
count, total, matched_df = count_main_nace_businesses(business_df, main_df)
print(f"{count} out of {total} businesses have a main NACE.")
```
