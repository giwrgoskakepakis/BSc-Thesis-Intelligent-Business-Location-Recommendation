## ***Load the Data***

```python
# Load the data
df = pd.read_csv(f"{base_path}{approach}Extracted CSV Files\\business_data_imputed.csv")
df.head()
```

## ***Find the top-10 locations per NACE***

```python
grouped = df.groupby(["NACE Code", "Neighborhood"])["Rating"].mean().reset_index()
```

```python
top10 = (
    grouped.sort_values(["NACE Code", "Rating"], ascending=[True, False])
    .groupby("NACE Code")
    .head(10)
)
```

```python
# Rank each community per NACE Code
top10["Rank"] = top10.groupby("NACE Code")["Rating"].rank(method="first", ascending=False)

# Pivot the table to wide format
pivoted = top10.pivot(index="NACE Code", columns="Rank", values="Neighborhood").reset_index()
```

```python
# Rename columns to Top1, Top2, Top3
pivoted.columns.name = None
pivoted = pivoted.rename(columns={1.0: "Top1", 2.0: "Top2", 3.0: "Top3", 4.0: "Top4", 5.0: "Top5", 6.0: "Top6", 7.0: "Top7", 8.0: "Top8", 9.0: "Top9", 10.0: "Top10"})
```

## ***Save the Data***

```python
pivoted.to_csv(f"{base_path}{approach}Extracted CSV Files\\top10.csv", index=False)
```
