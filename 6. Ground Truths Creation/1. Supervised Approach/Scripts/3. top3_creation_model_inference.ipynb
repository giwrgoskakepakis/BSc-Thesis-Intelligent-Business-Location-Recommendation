## ***Create the Top10***

```python
# Step 4: Extract top 3 communities per NACE
top10_df = (
    df.sort_values(by=['NACE Code', 'score'], ascending=[True, False])
      .groupby('NACE Code')
      .head(10)
      .groupby('NACE Code')['Neighborhood']
      .apply(list)
      .reset_index()
)
```

```python
# Step 5: Expand list into columns
top10_df[['Top1', 'Top2', 'Top3', 'Top4', 'Top5', 'Top6', 'Top7', 'Top8', 'Top9', 'Top10']] = pd.DataFrame(top10_df['Neighborhood'].tolist(), index=top10_df.index)
top10_df.drop(columns='Neighborhood', inplace=True)
```
