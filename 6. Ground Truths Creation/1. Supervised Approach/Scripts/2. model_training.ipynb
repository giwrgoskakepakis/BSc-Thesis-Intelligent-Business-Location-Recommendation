## ***Load the Data***

```python
df = pd.read_csv(f"{base_path}Extracted CSV Files\\ML_model_dataset.csv")
```

```python
print(df.shape)
print(df.columns)
df.head()
```

## ***Train the ML Model***

```python
nace_encoder = LabelEncoder()
df['NACE_Code_Encoded'] = nace_encoder.fit_transform(df['NACE Code'])
```

```python
# Step 1: Prepare feature set
non_feature_cols = ['NACE Code', 'Neighborhood', 'Municipal Community', 'ΠΕΡΙΦΕΡΕΙΑΚΗ ΕΝΟΤΗΤΑ', 'ΔΗΜΟΣ', 'ΔΗΜΟΤΙΚΗ ΕΝΟΤΗΤΑ', 'target']
feature_cols = [col for col in df.columns if col not in non_feature_cols]

X = df[feature_cols]
y = df['target']
```

```python
# Step 2: Train the classifier
pipeline = Pipeline([
    ('scaler', MinMaxScaler()),
    ('clf', RandomForestClassifier(n_estimators=100, random_state=42))
])
pipeline.fit(X, y)
```

```python
# Step 3: Predict probabilities for all rows (probability of class 1)
df['score'] = pipeline.predict_proba(X)[:, 1] 
```
