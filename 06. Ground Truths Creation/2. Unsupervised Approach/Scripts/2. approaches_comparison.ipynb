## ***Help Functions***

```python
# function that: scores the top3.csv on how close it is to the central neighborhoods
def central_score(df, top_cols, central_set):
    count = 0
    total = 0
    for col in top_cols:
        count += df[col].isin(central_set).sum()
        total += df[col].notnull().sum()
    return count, total, count / total if total > 0 else 0
```
