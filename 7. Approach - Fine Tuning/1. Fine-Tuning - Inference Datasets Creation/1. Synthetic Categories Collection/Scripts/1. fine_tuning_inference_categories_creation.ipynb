First of all, I will make 2 separate files that contain the 'Category' rows that will be used to create the Fine-Tuning and Inference datasets

## ***Initials***

```python
import pandas as pd
```

## ***Load the Data***

```python
df_main = pd.read_csv("C:\\Users\\Giorgos\\Desktop\\HMMY\\10ο Εξάμηνο\\Διπλωματική\\7. LLM - Fine Tuning\\1. Datasets Creation\\1. Synthetic Data Collection\\Extracted CSV Files\\Synthetic_Main_Categories.csv")
df_secondary = pd.read_csv("C:\\Users\\Giorgos\\Desktop\\HMMY\\10ο Εξάμηνο\\Διπλωματική\\7. LLM - Fine Tuning\\1. Datasets Creation\\1. Synthetic Data Collection\\Extracted CSV Files\\Synthetic_Secondary_Categories.csv")
```

## ***Create the Fine-Tuning Categories***

```python
# ---- Output 1: Fine-tuning categories ----

# From main: take first 3 categories
main_finetune = df_main[['NACE', 'Description', 'Category1', 'Category2', 'Category3']].copy()
main_finetune.columns = ['NACE', 'Description', 'Category1', 'Category2', 'Category3']

# From secondary: already fine-tune style
secondary_finetune = df_secondary[['NACE', 'Description', 'Category1', 'Category2', 'Category3']].copy()

# Concatenate them
all_finetune = pd.concat([main_finetune, secondary_finetune], ignore_index=True)
```

```python
print(all_finetune.shape)
all_finetune.head()
```

```python
# Save fine-tuning file
all_finetune.to_csv('C:\\Users\\Giorgos\\Desktop\\HMMY\\10ο Εξάμηνο\\Διπλωματική\\7. LLM - Fine Tuning\\1. Datasets Creation\\1. Synthetic Data Collection\\Extracted CSV Files\\Categories_Fine_Tuning.csv', index=False)
```

## ***Create the Inference Categories***
