## ***Initials***

```python
import matplotlib.pyplot as plt
import numpy as np
```

## ***Help Functions***

```python
def plot_metrics(attempts, metrics):

    # Plotting
    labels = list(metrics.keys())
    num_metrics = len(labels)
    num_attempts = len(attempts)

    x = np.arange(num_metrics)  # Metric positions on x-axis
    width = 0.25

    fig, ax = plt.subplots(figsize=(8, 5))

    # Plot each epoch's bar next to each other per metric
    for i in range(num_attempts):
        values = [metrics[m][i] for m in labels]
        ax.bar(x + (i - 1) * width, values, width, label=attempts[i])

    # Labels and formatting
    ax.set_ylabel('Score')
    ax.set_title('Comparison of Epoch Settings per Metric')
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.legend(title='Epochs')

    # Add value labels
    for i in range(num_attempts):
        values = [metrics[m][i] for m in labels]
        for j, val in enumerate(values):
            ax.text(x[j] + (i - 1) * width, val + 0.005, f'{val:.3f}', ha='center', fontsize=8)

    plt.tight_layout()
    plt.show()
```

## ***Visualize the Results***

```python
# Data
attempts = ['3 Epochs', '5 Epochs', '8 Epochs']

bert_precision = [0.9229, 0.9228, 0.9188]
bert_recall = [0.9133, 0.9202, 0.9179]
bert_f1 = [0.9181, 0.9215, 0.9183]
match_at_3 = [0.7862, 0.7799, 0.6981]
precision_at_3 = [0.3145, 0.3187, 0.2851]
recall_at_3 = [0.3145, 0.3187, 0.2851]

metrics = {
    'BERT Precision': bert_precision,
    'BERT Recall': bert_recall,
    'BERT F1': bert_f1,
    'Match@3': match_at_3,
    'Precision@3': precision_at_3,
    'Recall@3': recall_at_3
}
plot_metrics(attempts, metrics)
```
