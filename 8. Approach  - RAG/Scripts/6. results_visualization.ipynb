## ***Initials***

```python
import matplotlib.pyplot as plt
import numpy as np
```

## ***Help Functions***

```python
def plot_metrics(attempts, metrics):

    # Plotting
    labels = list(metrics.keys())
    num_metrics = len(labels)
    num_attempts = len(attempts)

    x = np.arange(num_metrics)  # Metric positions on x-axis
    width = 0.25

    fig, ax = plt.subplots(figsize=(8, 5))

    # Plot each epoch's bar next to each other per metric
    for i in range(num_attempts):
        values = [metrics[m][i] for m in labels]
        ax.bar(x + (i - 1) * width, values, width, label=attempts[i])

    # Labels and formatting
    ax.set_ylabel('Score')
    ax.set_title('Comparison of Justifications Number per Metric')
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.legend(title='Justifications')

    # Add value labels
    for i in range(num_attempts):
        values = [metrics[m][i] for m in labels]
        for j, val in enumerate(values):
            ax.text(x[j] + (i - 1) * width, val + 0.005, f'{val:.3f}', ha='center', fontsize=8)

    plt.tight_layout()
    plt.show()
```

## ***Visualize the Results***

```python
# Data
attempts = ['10 Justifications', '15 Justifications', '20 Justifications']

bert_precision = [0.8727, 0.8648, 0.8608]
bert_recall = [0.8381, 0.8314, 0.8290]
bert_f1 = [0.8550, 0.8477, 0.8446]
match_at_3 = [0.8302, 0.7862, 0.7107]
precision_at_3 = [0.3669, 0.3548, 0.2778]
recall_at_3 = [0.3669, 0.3522, 0.2893]

metrics = {
    'BERT Precision': bert_precision,
    'BERT Recall': bert_recall,
    'BERT F1': bert_f1,
    'Match@3': match_at_3,
    'Precision@3': precision_at_3,
    'Recall@3': recall_at_3
}
plot_metrics(attempts, metrics)
```
