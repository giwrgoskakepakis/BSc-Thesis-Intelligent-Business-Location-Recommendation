```python
# function that: creates an individual justification per feature of the neighborhood, and combines them 
# this time, we do not samle 4/7 fixed features, but we use all of them
# also, we do not use the feature importances for thr secondary features, 
def justify_neighborhood(neighborhood, neighborhood_stats, municipal_community_stats, neighborhood_colls, municipal_community_colls, all_features, fixed_features, max_reasons=5):
    
    individual_justifications = []
    used_features = set()


    # ================== Step 1: Justify fixed features ==================
    for feature in fixed_features:
        
        # Extract the stats of this feature
        if feature in neighborhood_colls:
            feature_stats = neighborhood_stats
        elif feature in municipal_community_colls:
            feature_stats = municipal_community_stats

        # Get the specific (FIXED) feature justification and add it to the total neighborhood justification
        feature_justification = justify_feature(feature, all_features[feature], feature_stats)
        if feature_justification:
            individual_justifications.append(feature_justification)
            used_features.add(feature)
        if len(individual_justifications) >= max_reasons:
            return f"{neighborhood} – {', '.join(individual_justifications)}."


    #  ================== Step 2: Add extra features (skipping any already used) ==================
    for feature in all_features:

        # Skip if the feature is not valid
        if feature in used_features:
            continue

        # Extract the stats of this feature
        if feature in neighborhood_colls:
            feature_stats = neighborhood_stats
        elif feature in municipal_community_colls:
            feature_stats = municipal_community_stats

        # Get the specific (EXTRA) feature justification and add it to the total neighborhood justification
        feature_justification = justify_feature(feature, all_features[feature], feature_stats)
        if feature_justification:
            individual_justifications.append(feature_justification)
            used_features.add(feature)
        if len(individual_justifications) >= max_reasons:
            break
    

    if not individual_justifications:
        return f"{neighborhood} – no significant distinguishing features found."
    return f"{neighborhood} – {', '.join(individual_justifications)}."
```

```python
# function that: returns stats for all the features, so each feature can later be compared later
def compute_features_stats(df, features):
    stats = {}
    for feature in features:
        stats[feature] = {
            "mean": df[feature].mean(),
            "q25": df[feature].quantile(0.25),
            "q50": df[feature].quantile(0.50),
            "q75": df[feature].quantile(0.75),
        }
    return stats
```

```python
# function that: returns the recommended features of a Neighborhood
def extract_neighborhood_features(neighborhood_name, neighborhood_ELSTAT_df, feature_cols):

    # find the 
    match1 = neighborhood_ELSTAT_df[neighborhood_ELSTAT_df['Neighborhood'] == neighborhood_name]
    if match1.empty:
        print(f"[WARNING] Neighborhood '{neighborhood_name}' not found in enriched dataset.")
        features = {}
    else:
        features = match1.iloc[0][feature_cols].to_dict()

    return features
```
