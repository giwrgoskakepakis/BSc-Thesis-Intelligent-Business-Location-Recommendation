# ***Department: ΥΠΗΡΕΣΙΩΝ***

```python
department="ΥΠΗΡΕΣΙΩΝ"
```

```python
df_real = pd.read_csv(f"/content/drive/My Drive/Diplomatiki/Success Probability - Real/data_QGIS_" +department +".csv")
print(f"Total samples: ", len(df_real))
df_real.head()
```

```python
df_pred = pd.read_csv(f"/content/drive/My Drive/Diplomatiki/Success Probability - Predicted/LLM_prediction_" +department +".csv")
print(f"Total samples: ", len(df_pred))
df_pred.head()
```

```python
df_real['predicted success probability'] = df_pred['success_probability']
df_real.rename(columns={'active years scaled': 'real success probability'}, inplace=True)
df_real.to_csv(f'merged_data_' + department + '.csv', index=False)
```
