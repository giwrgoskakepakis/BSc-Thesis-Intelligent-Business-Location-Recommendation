## ***Initials***

```python
import json
import glob
```

## ***Merge All the JSON Files***

Now, after i saved each API Call in a different .json file, i want to merge them all into 1 large file that contains all the Foursquare data

```python
all_places = []

for filename in glob.glob("places_*.json"):
    with open(filename, "r", encoding="utf-8") as file:
        places = json.load(file)
        all_places.extend(places)  # append the list of places from each file

# Save to one merged JSON file
with open("all_places_combined.json", "w", encoding="utf-8") as outfile:
    json.dump(all_places, outfile, ensure_ascii=False, indent=2)
```

```python
print(len(all_places))
```

```python
# Sanity check: count places in each file
total = 0

for filename in glob.glob("places_*.json"):
    with open(filename, "r", encoding="utf-8") as f:
        data = json.load(f)
        count = len(data)
        total += count
        print(f"{filename}: {count} places")

print(f"\nTotal across all files: {total} places")
```

## ***Remove the Duplicate Entries***

Since the grid circles are overlapping, i assume there are duplicates in the merged file, and therefore i have to remove them

```python
# open file
with open("all_places_combined.json", "r", encoding="utf-8") as f:
    all_places = json.load(f)
```

```python
# find unique places
unique_places = {}
for place in all_places:
    unique_places[place["fsq_id"]] = place

len(unique_places)
```

```python
# Convert back to a list
deduplicated_list = list(unique_places.values())

# Save to a new JSON file
with open("all_places_deduplicated.json", "w", encoding="utf-8") as f:
    json.dump(deduplicated_list, f, ensure_ascii=False, indent=2)
```

```python
print(f"Original count: {len(all_places)}")
print(f"Deduplicated count: {len(deduplicated_list)}")
```
