This file merged the individual .csv files that i got by manually performing web scraping multiple times in the site of Spitogatos about ***home rent prices***

## ***Initials***

```python
import pandas as pd
import glob
import os
```

## ***Load the Data***

```python
# Define the folder where the CSV files are located
folder_path = "C:/Users/Giorgos/Desktop/HMMY/10ο Εξάμηνο/Διπλωματική/Datasets/Web Scraping/Spitogatos/web scraps" 
```

```python
# Use glob to get all files that match the pattern
csv_files = glob.glob(os.path.join(folder_path, "SpitogatosVolosPage*.csv"))
```

## ***Merge the CSVs***

I performed multiple web scrapings MANUALLY about the ***home rent price***, not i want to merge them

```python
# Read and concatenate all CSVs
df_list = [pd.read_csv(file) for file in csv_files]
merged_df = pd.concat(df_list, ignore_index=True)
```

```python
merged_df = merged_df.drop(columns=['web-scraper-order', 'web-scraper-start-url', 'homeLinks-href', 'homeLinks'])
```

```python
merged_df['Price'] = merged_df['Price'].str.replace(" / μήνα", "", regex=False)
merged_df['Price'] = merged_df['Price'].str.replace("€", "", regex=False).str.strip()
merged_df['Price'] = pd.to_numeric(merged_df['Price'])
```
