## ***Initials***

```python
import matplotlib.pyplot as plt
import numpy as np
```

## ***Help Functions***

```python
def plot_metrics(attempts, metrics):

    # Plotting
    labels = list(metrics.keys())
    num_metrics = len(labels)
    num_attempts = len(attempts)

    x = np.arange(num_metrics)  # Metric positions on x-axis
    width = 0.25

    fig, ax = plt.subplots(figsize=(8, 5))

    # Plot each epoch's bar next to each other per metric
    for i in range(num_attempts):
        values = [metrics[m][i] for m in labels]
        ax.bar(x + (i - 1) * width, values, width, label=attempts[i])

    # Labels and formatting
    ax.set_ylabel('Score')
    ax.set_title('Comparison of Approaches')
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.legend(title='Approaches')

    # Add value labels
    for i in range(num_attempts):
        values = [metrics[m][i] for m in labels]
        for j, val in enumerate(values):
            ax.text(x[j] + (i - 1) * width, val + 0.005, f'{val:.3f}', ha='center', fontsize=8)

    plt.tight_layout()
    plt.show()
```

## ***Visualize the Results***

```python
# Data
attempts = ['Fine-Tuned LLM', 'RAG Pipeline']

bert_precision = [0.9228, 0.8727]
bert_recall = [0.9202, 0.8381]
bert_f1 = [0.9215, 0.8550]
match_at_3 = [0.7799, 0.8302]
precision_at_3 = [0.3187, 0.3669]
recall_at_3 = [0.3187, 0.3669]


metrics = {
    'BERT Precision': bert_precision,
    'BERT Recall': bert_recall,
    'BERT F1': bert_f1,
    'Match@3': match_at_3,
    'Precision@3': precision_at_3,
    'Recall@3': recall_at_3
}

plot_metrics(attempts, metrics)
```

```python
# Data
attempts = ['Fine-Tuned LLM', 'RAG Pipeline']

atpq = [37.56, 16.00]

metrics = {
    'Average Inference Time per Query (sec/iteration)': atpq,
}

plot_metrics(attempts, metrics)
```

```python
# Data
attempts = ['Fine-Tuned LLM', 'RAG Pipeline']
atpq = [37.56, 16.00]

# Plot
plt.figure(figsize=(6, 4))
bars = plt.bar(attempts, atpq, color=['skyblue', 'salmon'])
plt.ylabel('Seconds per Query')
plt.title('Average Inference Time per Query')

# Add values on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.2f}',
             ha='center', va='bottom')

plt.show()
```
